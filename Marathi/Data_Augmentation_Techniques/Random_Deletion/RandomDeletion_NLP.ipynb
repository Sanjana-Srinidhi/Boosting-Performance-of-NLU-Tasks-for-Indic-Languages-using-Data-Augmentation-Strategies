```python
# importing the required libraries
import pandas as pd
import random
```

```python
# reading the training dataset
train_df = pd.read_csv('/content/drive/MyDrive/NLP_Mini_Project/Data_Augmentation_Techniques/Back_Transilation/Train_clean.csv')
train_df.head()
```

```python
# defining function for augmentation
def kannada_augmentation(sentence, p = 0.1):
    words = sentence.split()
    new_words = []

    for word in words:
        if random.uniform(0, 1) < p:
            continue
        else:
            new_words.append(word)

    new_sentence = ' '.join(new_words)
    return new_sentence
```

```python
# appending to the training set
for index, row in train_df.iterrows():
  modified_sentence = kannada_augmentation(row['headline'])
  train_df.loc[len(train_df.index)] = [modified_sentence, row['label']]
```
