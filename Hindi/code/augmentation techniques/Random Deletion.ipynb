```python
# importing the required libraries
import pandas as pd
import random
```

```python
# reading the training dataset
train_df = pd.read_csv('/content/drive/MyDrive/NLP/data/clean-hindi-train.csv')
train_df = train_df[train_df.text.notnull()]
train_df['text'] = train_df['text'].str[:100]
train_df.head()
```

```python
# size of training set before augmentation
len(train_df)
```

```python
train_df.iloc[1]
```

```python
# defining function for augmentation
def random_deletion(sentence, p = 0.2):
    words = sentence.split()
    new_words = []

    for word in words:
        if random.uniform(0, 1) < p:
            continue
        else:
            new_words.append(word)

    new_sentence = ' '.join(new_words)
    return new_sentence
```

```python
dfaug = train_df
dfaug['text'] = dfaug['text'].apply(lambda x: random_deletion(x))
```

```python
dfaug.shape
```

```python
# Write updated DataFrame to output CSV file
train_df = train_df.append(dfaug, ignore_index=True)
```

```python
# shuffling the samples
train_df = train_df.sample(frac = 1, ignore_index = True)
train_df.head()
```

```python
# writing the augmented result to a csv file
train_df.to_csv('RandomDeletionAugData.csv', index=False)
```
